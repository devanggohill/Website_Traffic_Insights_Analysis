```python
df = pd.read_csv("data-export.csv", skiprows=1)


df.head()
```

```python
print("Column Names:\n", df.columns.tolist())

for col in df.select_dtypes(include='object').columns:
    print(f"\nUnique values in '{col}':")
    print(df[col].unique()[:10])
```

```python
df.rename(columns={
    'Session_primary_channel_group_Default_channel': 'Channel_Group',
    'Date_hour_YYYYMMDDHH': 'Date_Hour',
    'Average_engagement_time_per_session': 'Avg_Engagement_Time',
    'Engaged_sessions_per_user': 'Engaged_Sessions_per_User',
    'Engagement_rate': 'Engagement_Rate',
    'Event_count': 'Event_Count'
}, inplace=True)


print("Null values per column:\n", df.isnull().sum())

print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Cell 9: Rename columns properly
df.rename(columns={
    'Session primary channel group (Default channel group)': 'Channel_Group',
    'Date + hour (YYYYMMDDHH)': 'Date_Hour',
    'Engaged sessions': 'Engaged_Sessions',
    'Average engagement time per session': 'Avg_Engagement_Time',
    'Engaged sessions per user': 'Engaged_Sessions_per_User',
    'Events per session': 'Events_per_Session',
    'Engagement rate': 'Engagement_Rate',
    'Event count': 'Event_Count'
}, inplace=True)

# Convert numeric columns safely
numeric_cols = [
    'Users', 'Sessions', 'Engaged_Sessions', 'Avg_Engagement_Time',
    'Engaged_Sessions_per_User', 'Events_per_Session', 'Engagement_Rate', 'Event_Count'
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Confirm dtypes
df.dtypes
```

```python
# Question 1 (Improved): Daily trend of Users and Sessions


daily_df = df.resample('D', on='Date_Hour').agg({'Users': 'sum', 'Sessions': 'sum'})

plt.figure(figsize=(12, 6))
plt.plot(daily_df.index, daily_df['Users'], label='Daily Users', marker='o')
plt.plot(daily_df.index, daily_df['Sessions'], label='Daily Sessions', marker='x')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Daily Users vs Sessions Trend')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  Question 2: Which marketing channel brought the highest number of users?


channel_users = df.groupby('Channel_Group')['Users'].sum().sort_values(ascending=False)


plt.figure(figsize=(10,6))
channel_users.plot(kind='bar', color='skyblue')
plt.title('Top Channels by Total Users')
plt.ylabel('Total Users')
plt.xlabel('Marketing Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display values
print(channel_users)
```

```python
# Question 3: Which channel has the highest average engagement time?


channel_engagement = df.groupby('Channel_Group')['Avg_Engagement_Time'].mean().sort_values(ascending=False)


plt.figure(figsize=(10,6))
channel_engagement.plot(kind='bar', color='lightgreen')
plt.title('Average Engagement Time by Channel')
plt.ylabel('Avg Engagement Time (seconds)')
plt.xlabel('Marketing Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


print(channel_engagement)
```

```python
# ðŸ“Š Question 4: How does engagement rate vary across different traffic channels?

channel_engagement_rate = df.groupby('Channel_Group')['Engagement_Rate'].mean().sort_values(ascending=False)


plt.figure(figsize=(10,6))
channel_engagement_rate.plot(kind='bar', color='orange')
plt.title('Average Engagement Rate by Channel')
plt.ylabel('Engagement Rate')
plt.xlabel('Marketing Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(channel_engagement_rate)
```

```python
# Question 5: Which channels drive more engaged sessions vs non-engaged?


engagement_comparison = df.groupby('Channel_Group')[['Sessions', 'Engaged_Sessions']].sum()


engagement_comparison['Engaged_Session_Rate'] = (
    engagement_comparison['Engaged_Sessions'] / engagement_comparison['Sessions']
)


engagement_comparison = engagement_comparison.sort_values('Engaged_Session_Rate', ascending=False)


plt.figure(figsize=(10,6))
engagement_comparison['Engaged_Session_Rate'].plot(kind='bar', color='purple')
plt.title('Engaged Sessions Rate by Channel')
plt.ylabel('Engaged Sessions / Total Sessions')
plt.xlabel('Marketing Channel')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


print(engagement_comparison[['Engaged_Session_Rate']])
```

```python
# Question 6: What time periods see the most user activity?


df['Hour'] = df['Date_Hour'].dt.hour
df['DayOfWeek'] = df['Date_Hour'].dt.day_name()


hourly_users = df.groupby('Hour')['Users'].sum()


daily_users = df.groupby('DayOfWeek')['Users'].sum()


days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_users = daily_users.reindex(days_order)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
hourly_users.plot(kind='bar', color='steelblue')
plt.title('Total Users by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Users')


plt.subplot(1, 2, 2)
daily_users.plot(kind='bar', color='teal')
plt.title('Total Users by Day of Week')
plt.xlabel('Day')
plt.ylabel('Users')

plt.tight_layout()
plt.show()
```

```python
# Question 7: How effective are website events in driving user engagement?

event_analysis = df.groupby('Channel_Group')[['Event_Count', 'Events_per_Session']].mean().sort_values('Events_per_Session', ascending=False)

# Plot events per session
plt.figure(figsize=(10,6))
event_analysis['Events_per_Session'].plot(kind='bar', color='coral')
plt.title('Average Events per Session by Channel')
plt.xlabel('Channel')
plt.ylabel('Events per Session')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Show table
print(event_analysis)
```
